```python
text = """Heyyy!!! 👋 I'm sooo excited about this AI/ML course!!! Visit https://example.com NOW!!! 😎😎
I’ve completed 2 assignments, but honestly... I didn’t understand 100% of it 😅 — can you pls help me???
Email me @ mailto:anusha123@gmail.com ASAP!!! #learning #AI #2026 🚀🚀"""
```

```python
#lowercasing
lower_text = text.lower()
lower_text
```

```python
#remove email
cleaned = re.sub(r'\S+@\S+','',lower_text)
cleaned
```

```python
#remove urls

cleaned = re.sub(r'http\S+','',cleaned)
cleaned
```

```python
#handling punctuation
import re

cleaned = re.sub(r'[^\w\s]','',cleaned)
cleaned
```

```python
#numerical
cleaned = ''.join(c for c in cleaned if not c.isdigit())
cleaned
```

```python
cleaned = re.sub(r'\s+', ' ', cleaned).strip()
print(cleaned)
```

```python

import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(cleaned)
filtered = [token.text for token in doc if not token.is_stop]
print(filtered)
```

```python
text = " ".join(filtered)
print(text)
```
