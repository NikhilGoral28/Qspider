```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
items= [
    ['Laptop', 'Mouse'],
    ['Keyboard', 'Mouse', 'Headphones'],
    ['Laptop', 'Keyboard'],
    ['Laptop', 'Mouse', 'Keyboard'],
    ['Monitor', 'Keyboard'],
    ['Laptop', 'Monitor', 'Mouse'],
    ['Mouse', 'USB Drive'],
    ['Laptop', 'Headphones'],
    ['Keyboard', 'Mouse', 'USB Drive'],
    ['Laptop', 'Mouse', 'Monitor']
]
items
```

```python
t_encoder = TransactionEncoder()
ecoded_items = t_encoder.fit_transform(items)

feature_names = t_encoder.get_feature_names_out()
feature_names
```

```python
df = pd.DataFrame(ecoded_items,columns=feature_names)
df
```

```python
frequent_pattern = fpgrowth(df,min_support=0.4,use_colnames=True)
frequent_pattern
```

```python
rules = association_rules(
    frequent_pattern,
    metric='confidence',
    min_threshold=0.5
)

rules
```

```python
a = rules.iloc[:,[0,1,4,5]]
a
```
