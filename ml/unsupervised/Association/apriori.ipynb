```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
items = [
    ["bread","milk","butter","jam"],
    ["bread","butter","milk"],
    ["bread","juice","curd"],
    ["bread","milk","juice"],
    ["butter","milk","juice"]
]
items
```

```python
t_encoder = TransactionEncoder()
ecoded_items = t_encoder.fit_transform(items)

feature_names = t_encoder.get_feature_names_out()
feature_names
```

```python
df = pd.DataFrame(ecoded_items,columns=feature_names)
df
```

```python
frequent_pattern = apriori(df,min_support=0.5,use_colnames=True)
frequent_pattern
```

```python
x = association_rules(frequent_pattern,metric='confidence',min_threshold=0.7)
x
```

```python
a = x.iloc[:,[0,1,4,5]]
a
```
