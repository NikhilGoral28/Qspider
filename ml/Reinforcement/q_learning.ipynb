```python
# Build an agent that learns to reach the goal
# problem:  moving left or right to reach the goal 5 (0,1,2,3,4,5)
# states: 0,1,2,3,4,5

#always the state starts at 0
#goal is to reach the 5

#ACTIONS  :    we have two actions to perform
# we are assuming moving left-0 and moving right-1
# action:0 --->left
# action:1 --->right

# REWARDS
# Reaching the goal: +10
# every other step(wrong): -1 
```

```python
import numpy as np
import random

states = 6
actions = 2

Q = np.zeros((states,actions))

alpha=0.05 #learning rate
gamma=0.9 #  Importance of future rewards
          # 0.9 → future rewards are important
epsilon = 0.2
# exploration rate--when the agent should perform the action or learn from the past data
# Probability of taking random action
# 0.2 → 20% explore, 80% exploit
episodes=500 #no. of times the game played inorder to get good result


for i in range(episodes):
  state=0 #always the state will start from 0
  while state!=5:  #loop continues until the agent reaches the goal
    if random.uniform(0,1)<epsilon:
      action=random.choice([0,1])    #makes an action(it chooses randomly 0 or 1)
    else:
      action=np.argmax(Q[state])  #learns from the Q-table

    next_state=state-1 if action==0 else state+1  #If 0 → move left
                                                  # If 1 → move right

    next_state=max(0,min(next_state,5)) #agent moves within the boundary (0-5)

    reward=10 if next_state==5 else -1  #rewards and penalty

    #Updates the quality value in the table
    Q [state,action]=Q[state,action]+alpha*(reward+gamma*np.max
                                              (Q[next_state])-Q[state,action])
    # Agent learns:
    # Was this action better or worse than expected?

    state=next_state#Update current position


print(Q)
print(i)
```
