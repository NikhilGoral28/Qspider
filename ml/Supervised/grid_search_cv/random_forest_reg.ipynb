```python
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data1 = fetch_california_housing()
data1
```

```python
df1 = pd.DataFrame(data1.data,columns=data1.feature_names)
```

```python
df1['Result'] = data1.target
```

```python
sc = StandardScaler()
d1 = sc.fit_transform(df1)
d1
```

```python
df2 = pd.DataFrame(d1,columns=df1.columns)
df2
```

```python
x1  = df2.drop(columns='Result',axis=1)
y1 = df2['Result']
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1)
```

```python
random_forest_reg = RandomForestRegressor()
```

```python
params = {
    'n_estimators': [150,180,200,210],
    'max_depth': [10,15,20,25]
}
```

```python
grid = GridSearchCV(estimator=random_forest_reg,param_grid=params)
grid.fit(x_train1,y_train1)
```

```python
pred1 = grid.predict(x_test1)
```

```python
print("Training eff: ", grid.score(x_train1,y_train1))
print("testing eff: ", grid.score(x_test1,y_test1))
```

```python
r2_score(y_test1,pred1)
```

```python
mean_absolute_error(y_test1,pred1)
```

```python
mean_squared_error(y_test1,pred1)
```
