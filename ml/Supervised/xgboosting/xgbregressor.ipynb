```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from xgboost import XGBRFRegressor
```

```python
df = pd.read_csv('D:\Qspider\ml\gradient_boosting\student_exam_scores.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['student_id'],axis=1)
```

```python
x = df.drop(columns='exam_score',axis=1)
y = df['exam_score']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xg = XGBRFRegressor()
```

```python
xg.fit(x_train,y_train)
```

```python
pred = xg.predict(x_test)
pred
```

```python
print('training score: ',xg.score(x_train,y_train))
print('testing score: ',xg.score(x_test,y_test))
```

```python
r2_score(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```
