```python
from sklearn import datasets
import pandas as pd
dir(datasets)
```

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
data
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
df["MedHouseVal"] = data.target
```

```python
df
```

# # Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
d1 = sc.fit_transform(df)
d1
```

```python
df = pd.DataFrame(d1,columns=df.columns)
df
```

```python
x = df.drop("MedHouseVal",axis=1)
y = df["MedHouseVal"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import  train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
x_train,y_train
```

```python
from sklearn.linear_model import LinearRegression

l = LinearRegression()
l
```

```python
l.fit(x_train,y_train)
```

```python
#prediction

pred = l.predict(x_test)
pred
```

```python
#score
print("Training Eff: ", l.score(x_train,y_train))
print("Testing Eff: ", l.score(x_test,y_test))
```

```python
#metrics

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print("r2_score: ", r2_score(y_test,pred))
print("Mean_squared error: ", mean_squared_error(y_test,pred))
print("mean_absolute_error: ",mean_absolute_error(y_test,pred))
```

```python
#regularazation 

from sklearn.linear_model import Lasso,Ridge

lasso = Lasso(alpha=0.005)
lasso.fit(x_train,y_train)
```

```python
print("Training Eff: ", lasso.score(x_train,y_train))
print("Testing Eff: ", lasso.score(x_test,y_test))
```

```python
print("Training Eff: ", ridge.score(x_train,y_train))
print("Testing Eff: ", ridge.score(x_test,y_test))
```
