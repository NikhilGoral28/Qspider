```python
from sklearn.datasets import fetch_california_housing,load_wine
from sklearn.ensemble import AdaBoostClassifier,AdaBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score,f1_score,r2_score,mean_absolute_error,mean_squared_error
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
california_housing = fetch_california_housing()
california_housing
```

```python
df = pd.DataFrame(california_housing.data,columns=california_housing.feature_names)
```

```python
df
```

```python
df['Result'] = california_housing.target
```

```python
df.head()
```

```python
x = df.drop(columns='Result',axis=1)
y = df['Result']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
#y_train.shape
```

```python
dt = DecisionTreeRegressor(max_depth=1)
```

```python
ad = AdaBoostRegressor(estimator=dt)
```

```python
ad
```

```python
ad.fit(x_train,y_train)
```

```python
pred = ad.predict(x_test)
```

```python
print("Training Score: ", ad.score(x_train,y_train))
print("Testing Score: ", ad.score(x_test,y_test))
```

```python
r2_score(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
## classiication

wine = load_wine()
```

```python
df1 = pd.DataFrame(wine.data,columns=wine.feature_names)
```

```python
df1
```

```python
df1["Result"] = wine.target
```

```python
df1.head()
```

```python
x1 = df1.drop(columns='Result',axis=1)
y1 = df1["Result"]
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
dtc = DecisionTreeClassifier(max_depth=1)
```

```python
adc = AdaBoostClassifier(estimator=dtc,n_estimators=100)
```

```python
adc
```

```python
adc.fit(x_train1,y_train1)
```

```python
pred1 = adc.predict(x_test1)
```

```python
confusion_matrix(y_test1,pred1)
```

```python
accuracy_score(y_test1,pred1)
```

```python
precision_score(y_test1,pred1,average='weighted')
```

```python
recall_score(y_test1,pred1,average='weighted')
```

```python
f1_score(y_test1,pred1,average='weighted')
```
