```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,precision_score,f1_score,recall_score
```

```python
data = load_wine()
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df['result'] = data.target
```

```python
df.head()
```

```python
df['result'].nunique()
```

```python
x = df.drop('result',axis=1)
y = df['result']
```

```python
#split the data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
#sccaling
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
```

```python
x_test_scaled = sc.transform(x_test)
```

```python
#model building
dt = DecisionTreeClassifier(max_depth=5,criterion='gini')
dt.fit(x_train_scaled,y_train)
```

```python
pred = dt.predict(x_test_scaled)
```

```python
accuracy_score(y_test,pred)
```

```python
classification_report(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
precision_score(
    y_test,
    pred,
    average='macro'
)
```

```python
recall_score(   y_test,
    pred,
    average='macro')
```

```python
f1_score( y_test,
    pred,
    average='macro')
```
