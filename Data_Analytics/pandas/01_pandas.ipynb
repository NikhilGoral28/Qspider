```python
import pandas as pd
import numpy as np
```

## Series

```python
#for string
pd.Series('ramu')
```

```python
pd.Series('ramu',index=[0],name='age')
```

```python
#for list
pd.Series([10,20,30,40,50])
```

```python
pd.Series([10,20,30,40,50],index=np.arange(0,5),name='age')
```

```python
#for tuple
pd.Series((10,20,30,40,50))
```

```python
pd.Series((10,20,30,40,50),index=np.arange(0,5),name='marks')
```

```python
#for dict

import pandas as pd

pd.Series({'a':1,'b':2})
```
