```python
import numpy as np
n = 30000
np.random.seed(42)
```

```python
Name = ['stud_'+str(i).zfill(5) for i in range(1,n+1)]
ph = np.random.randint(6666666666,9999999999,n,dtype=np.int64)

email = [i+'@gmail.com' for i in Name]
type = np.random.choice(['INC','PAID'],n)
```

```python
tenth_per = np.round(np.random.randint(66,99,n)+np.random.rand(n),2)
tenth_YOP = np.random.choice([2020,2019,2018],n)
```

```python
twelth_per = np.round(np.random.randint(66,99,n)+np.random.rand(n),2)
twelth_YOP = [(i+2).item() for i in tenth_YOP]
```

```python
print(tenth_YOP)
print(twelth_YOP)
```

```python
stream = np.random.choice(['BCA','BCS','BE','BTECH'],n)
```

```python
degree_per = np.round(np.random.randint(66,99,n)+np.random.rand(n),2)
```

```python
degree_YOP = []
for i,year in enumerate(twelth_YOP):
    if stream[i] == "BE" or stream[i] == 'BTECH':
        degree_YOP.append(year+4)
    else:
        degree_YOP.append(year+3)
```

```python
print(stream)
print(twelth_YOP)
print(degree_YOP)
```

```python
mas_stream = np.random.choice(['MCA','ME','MTECH',None],n)
mas_stream
```

```python
mas_per = []
for i in mas_stream:
    if i != None:
        mas_per.append(np.random.randint(37,99))
    else:
        mas_per.append(np.nan)

mas_per = np.array(mas_per)
mas_per.size
```

```python
mas_yop = []
for i,j in zip(mas_stream,degree_YOP):
    if i!= None:
        mas_yop.append(j+2)
    else:
        mas_yop.append(np.nan)

mas_yop = np.array(mas_yop)
mas_yop.size
```

```python
course = np.random.choice(['PFD','PDA','PDS','PS'],n)
course
```

```python
status = np.random.choice(['Placed','Not placed'],n)
status
```

```python
import pandas as pd

Py_df = pd.DataFrame({
    'Name':Name,
    'Phone Number':ph,
    'Email ID':email,
    'Type':type,
    'course':course,
    'Tenth YOP':tenth_YOP,
    'Tenth Percentage':tenth_per,
    'PUC YOP':twelth_YOP,
    'PUC Percentage':twelth_per,
    'Degree YOP':degree_YOP,
    'Degree Stream':stream,
    'Degree Percentage':degree_per,
    'Masters YOP':mas_yop,
    'Masters Stream': mas_stream,
    'Masters Percentage': mas_per})
```
