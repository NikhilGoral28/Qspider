```python
def palindrome(num,rev=0):
    if num == 0:
        print(f'Execution of {num}')
        print(f"reverse at {num} is {rev}")
        return rev
    else:
        print(f'Execution of {num}')
        print(f"reverse at {num} is {rev}")
        print()
        return palindrome(num//10,rev*10+num%10)



num = 121
print(num == palindrome(num))
```

```python
#count digit using recursion

def count_digit(num):
    if num == 0:
        print(f'number is {num}')
        return 0
    else:
        print(f'number is {num}')
        return 1 + count_digit(num//10)
    
print(count_digit(12345))
```

```python
#factorial of a number

def fact(num):
    if num == 0:
        print(f'num is {num}')
        return 1
    else:
        print(f'num is {num}')
        return num * fact(num-1)
    

print(fact(5))
```

```python
#sum of first 10 

def total(num):
    if num == 0:
        print(f"num is {num}")
        return 0
    else:
        print(f"num is {num}")
        return num + total(num-1)
    
print(total(10))
```

```python
#sum of digit

def sumDigit(num):
    if num == 0:
        print(f"num is {num}")
        print(f"num%10 for the num is {num%10}")
        return 0
    else:
        print(f"num is {num}")
        print(f"num%10 for the num is {num%10}")
        return num%10 + sumDigit(num//10)

print(sumDigit(123))
```
