## get method

```python
#get method

data = {"ename":"smith","sal":98000}

print(data.get('sal'))
```

```python
print(data.get('job'))
```

```python
print(data.get('job','analyst'))
```

```python
print(data)
```

## update method

```python
data = {"ename":"smith","sal":98000}
data.update({"job":"analyst"})

print(data)
```

```python
data = {"ename":"smith","sal":98000}
data.update({"sal":99000})
print(data)
```

```python
data.update(sal=100000)
print(data)
```

## items method

```python
data = {"ename":"smith","sal":98000}
data.items()
```

## key method

```python
data = {"ename":"smith","sal":98000}
data.keys()
```

## values method

```python
data = {"ename":"smith","sal":98000}
data.values()
```

## pop method

```python
data = {"ename":"smith","sal":98000}
data.pop('sal')  # removes and return value of the specified key
```

```python
print(data)
```

## popitem method

```python
#remove the last inserted key value pair

data = {"ename":"smith","sal":98000}
data.popitem()
```

```python
print(data)
```

## clear method

```python
#clear everything in the dict

data = {"ename":"smith","sal":98000}
data.clear()
```

```python
print(data)
```
