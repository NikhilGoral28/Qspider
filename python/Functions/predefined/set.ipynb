# add method

```python
#add methods add the single element to the set

s1 = {'smith','martin'}

s1.add(65)
print(s1)
```

```python
s1 = {'smith','martin',65}

s1.add(65)
print(s1)
```

# remove method

```python
#used to remove the specified element from the set

s1 = {'smith','martin',65}

s1.remove(65)
print(s1)
```

```python
#if element not found gives the error

s1 = {'smith','martin'}

s1.remove(65)
print(s1)
```

# update method

```python
#used to add multiple elements at a time

s1 = {'smith','martin'}

s1.update({65,32,23})
print(s1)
```

```python
s1 = {'smith','martin'}

s1.update([65,32,23])
print(s1)
```

# pop method

```python
# used to remove the ramdom element or first apeared element it will remove

s1 = {'smith','martin'}

s1.pop()
```

```python
print(s1)
```

# clear method

```python
#used to  cleat the set

s1 = {'smith','martin'}

s1.clear()
print(s1)
```

```python
s2 = set()
s2.clear()

print(s2)
```

# union method

```python
# used to display unmatched elements from the both set

s1 = {1,2,3,4}
s2 = {3,4,5,6,7}

result = s1.union(s2)

print(result)
```
