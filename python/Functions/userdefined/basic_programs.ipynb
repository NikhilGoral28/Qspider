```python
#reverse a number

def reverseNum(num):
    rev = 0
    while num>0:
        rev = rev*10 + (num%10)
        num //= 10
    
    return rev

print(reverseNum(123))
```

```python
#sum of digit of a num

def digitSum(num):
    total = 0

    while num > 0:
        total += (num%10)
        num //= 10

    return total

print(digitSum(123))
```

```python
#num is palindrome or not

def checkPalindrome(num):
    rev = 0
    temp = num

    while temp > 0:
        rev = rev * 10 + (temp%10)
        temp //= 10

    if rev == num:
        return f"{num} is palindrome"
    else:
        return f"{num} is not palindrome"
    
print(checkPalindrome(121))
```

```python
#find the largest number of two numbers

def findLargest(num1,num2):

    if num1 > num2:
        return f'{num1} is largest'
    elif num2 > num1:
        return f"{num2} is largest"
    
    else:
        return "both equal"
    

print(findLargest(5,6))
```

```python
# check the number is armstrong number

def checkArmstrong(num):
    length = len(str(num))
    temp = num
    arm = 0

    while temp > 0:
        arm += (temp%10)**length
        temp //= 10

    if arm == num:
        return f"{num} is armstrong number"
    
    else:
        return f"{num} is not a armstrong number"
    

print(checkArmstrong(123))
```

```python
#check num is strong number is not

def CheckStrong(num):
    temp = num
    strong = 0

    while temp > 0:
        digit = temp %10
        fact = 1

        for i in range(1,digit+1):
            fact *= i

        strong += i

        temp //= 10
    
    if strong == num:
        return f"{num} is strong number"
    else:
        return f"{num} is not a strong number"

print(CheckStrong(1234))
```

```python
#check a perfect number

def Perfectnumber(num):
    total = 0

    for i in range(1,num):
        if num % i == 0:
            total += i
    
    if total == num:
        return f"{num} is a perfect number"
    else:
        return f"{num} is not a perfect number"
    
print(Perfectnumber(6))
```

```python
#fibonnacci of n term

def fibo(n):
    a,b=0,1

    for i in range(n):
        print(a,end=' ')
        a,b = b,a+b

fibo(8)
```

```python
#count of digits

def digitcounter(num):
    count = 0
    while num > 0:
        count += 1
        num //= 10

    return count

print(digitcounter(1234))
```

```python
#sum of starting natural numbers

def SumOfNaturalNum(num):
    totol = 0
    for i in range(1,num+1):
        totol += i

    return totol

print(SumOfNaturalNum(5))
```

```python
#find power of a number

def PowOfNum(num,n):
    return num**n

print(PowOfNum(2,3))
```

```python
#check number is neon or not

def CheckNeon(num):
    temp = num
    squared = num**2
    Neon = 0

    while squared > 0:
        Neon += (squared%10)
        squared //= 10

    if Neon == num:
        return f"{num} is neon number"
    else:
        return f"{num} is not neon number"

print(CheckNeon(9))
```

```python
#num is spy number or not


def CheckSpy(num):
    pro = 1
    total = 0
    temp = num

    while num > 0:
        pro *= (num%10)
        total += (num%10)

        num //= 10
    
    if pro == total:
        return f"{temp} is spy number"
    else:
        return f"{temp} is not a spy number"
    
print(CheckSpy(6))
```
